```python
a = 9.45676
b = 10.9289
grid_size = np.array([a, b])
nx = 10
ny = 10
ngrid = nx * ny
x = np.linspace(0, grid_size[0], nx)
y = np.linspace(0, grid_size[1], ny)
xx,yy=np.meshgrid(x, y)

xx_reshape=np.reshape(xx,(ngrid,1))
yy_reshape=np.reshape(yy,(ngrid,1))
virtual_atoms=np.array(np.column_stack((xx_reshape, yy_reshape,)))

plt.plot(xx, yy, marker='.', color='k', linestyle='none')
```

```python
# Read the input file
with open('mos2_slab_coordinates', 'r') as file:
    lines = file.readlines()

# Initialize an empty list to store atomic positions
atomic_positions = []

# Loop through the lines in the file
for line in lines:
    # Split each line into tokens (columns) using whitespace as the delimiter
    tokens = line.strip().split()
    
    # Check if there are at least 4 tokens (element symbol and x, y, z coordinates)
    if len(tokens) >= 4:
        # Extract the element symbol, and convert the x, y, z coordinates to float values
        element_symbol = tokens[0]
        x, y, z = map(float, tokens[1:4])
        
        # Append the element symbol and coordinates to the atomic_positions list
        atomic_positions.append([element_symbol, x, y, z])

# Convert the atomic_positions list to a NumPy array
mos2_pos = np.array(atomic_positions)
#print(mos2_pos)
# Print the resulting array
for entry in mos2_pos:
    element_symbol, x, y, z = entry
    print( element_symbol, x, y, z)
```

```python
df_pos=pd.DataFrame(mos2_pos,columns=['Symbol', 'x','y','z'])
print(df_pos)
```

```python
vdw_radii = {
    'Mo': 2.0, 
    'S': 1.85,
    'O': 1.52,
    'H': 1.2,
}

df_pos['x'] = pd.to_numeric(df_pos['x'])
df_pos['y'] = pd.to_numeric(df_pos['y'])

def calculate_z_coordinate(grid_point, atom_data, vdw_radii):
    x, y = grid_point[:2] 
    min_distance = float("1.5") # 1.4 - 1.5 is the bond length between S and O, 
    z_coordinate = None

    for i, row in atom_data.iterrows():
        distance = np.linalg.norm(np.array([x, y]) - np.array([row["x"], row["y"]]))
        vdw_radius = vdw_radii.get(row["Symbol"], 0.0)
        vdw_sum = vdw_radius  # Use the VDW radius of the corresponding element
        if distance > min_distance and distance >= vdw_sum:
            min_distance = distance
            z_coordinate = float(row["z"])
        if z_coordinate is None:
            z_coordinate = 5.4
        else:
            if z_coordinate < 1.2:
                z_coordinate += 4.0  
            elif 1.2 <= z_coordinate < 2.5:
                z_coordinate += 3.5
            elif 2.5 <= z_coordinate < 3.5:
                z_coordinate += 2.5
            elif 3.2 <= z_coordinate < 4.0:
                z_coordinate += 1.5
            elif 4.0 <= z_coordinate <= 4.8:
                z_coordinate += 0.7
    return z_coordinate

# Place adsorbates on the grid with Z-coordinates close to the nearest atoms and considering van der Waals radii
adsorbate_coordinates = []
for grid_point in virtual_atoms:
    x, y = grid_point[:2]  # Extract (x, y) coordinates
    z = calculate_z_coordinate(grid_point, df_pos, vdw_radii)
    adsorbate_coordinates.append((x, y, z))
# Print the resulting adsorbate coordinates
for coordinate in adsorbate_coordinates:
    #print(coordinate)

    print(np.array(coordinate))
```

```python
x_coordinates, y_coordinates, z_coordinates = zip(*adsorbate_coordinates)
# Reshape x, y, and z coordinates into 1D arrays
x_reshaped = np.array(x_coordinates).reshape(-1)
y_reshaped = np.array(y_coordinates).reshape(-1)
z_reshaped = np.array(z_coordinates).reshape(-1)
# Create a dictionary to build the DataFrame
data = {'x': x_reshaped, 'y': y_reshaped, 'z': z_reshaped}
# Create the DataFrame
grid_3_d = pd.DataFrame(data)

grid_3_d.head()
```
